# Prob of Emergence for Mask Model

$P(\text{wear a mask}) = m$

$E_{L, 0}$ := The epidemic process going through node v who is **active** at level $L$ given he **wear** a mask will be extinct 

$E_{L, 1}$ := The epidemic process going through node v who is **active** at level $L$ given he **doesn't wear** a mask will be extinct 


## 1. Condition on num of neighbors from lower level






$B_k$ := Node v has $k - 1$ lower level neighbors

$P(E_{L, i}) = \sum_{k = 1}^{\infty}{P(E_{L, i} | B_k) * P(B_k)}$

where

$i = 0,1$

$P(B_k) = \frac{k p_{k}}{\langle k\rangle}$

## 2. Condition on num of neighbors wearing a mask

$N$ := Num of neighbors wearing masks 
 
$N \sim \operatorname{Binomial}\left(k - 1, m\right)$


$ P(E_{L, i} | B_k) = \sum_{n = 0}^{k - 1} P(E_{L, i} | B_k, N = n) * \left(\begin{array}{c}
k - 1 \\
n
\end{array}\right) * m^{n} * (1 - m)^{k - 1 - n} $

## 3. Condition on num of active lower level neighbors

$I_0$ := $I_0$ of $k-1$ lower level neighbors active and wear masks and **the process will extinct.**

$I_1$ := $I_1$ of $k-1$ lower level neighbors active and don't wear masks and **the process will extinct.**

$I_0 \sim \operatorname{Binomial}\left(n, P(E_{L-1,0})\right)$

$I_1 \sim \operatorname{Binomial}\left(k - 1 - n, P(E_{L-1,1})\right)$


$P(E_{L, i} | B_k, N = n) = \sum_{k_0 = 0}^{n}\sum_{k_1 = 0}^{k-1-n}P(E_{L, i} | B_k, N = n, I_0 = k_0, I_1 = k_1) * \left(\begin{array}{c}
n \\
k_0
\end{array}\right) \left(\begin{array}{c}
k - 1 - n \\
k_1
\end{array}\right) * P(E_{L-1, 0})^{k_0} * (1 - P(E_{L-1,0}))^{n - k_0} * P(E_{L-1, 1})^{k_1} * (1 - P(E_{L-1, 1}))^{k - 1 - n - k_1}$

where

$i = 0,1$

## 4. $P(E_{L, i} | B_k, N = n, I_0 = k_0, I_1 = k_1)$

$R$ := $B_k, N = n, I_0 = k_0, I_1 = k_1$

### 4.1 $P(E_{L,0} | R)$



$P(E_{L,0} | R)$

$ = P(\text{At least 1 of node v's active neighbors succssfully infects it, given node v wears a mask} | R)$

$ = 1 - (1 - T_2)^{k_0} * (1 - T_4)^{k_1}$

### 4.1 $P(A_{L,1} | R)$



$P(A_{L,1} | R)$

$ = P(\text{At least 1 of node v's active neighbors succssfully infects it, given node v doesn't a mask} | R)$

$ = 1 - (1 - T_1)^{k_0} * (1 - T_3)^{k_1}$
